# 3) Создайте две numpy матрицы размера nхn и заполните их случайными значениями.

```python
matrix_n = 300
matrix_A = np.random.randint(0, 10, size=(matrix_n, matrix_n))
matrix_B = np.random.randint(0, 10, size=(matrix_n, matrix_n))
```

Посчитайте их поэлементное и матричное произведения. Для этого реализуйте функции:
1. mult – матричное перемножение без использования numpy
2. el_mult – поэлементное перемножение без использования numpy
3. np_mult – матричное перемножение c использованием numpy
4. np_el_mult – поэлементное перемножение с использованием numpy

```python
def mult(matrix1, matrix2):
    n = len(matrix1)
    result_matrix = [[0 for _ in range(n)] for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                result_matrix[i][j] += matrix1[i][k] * matrix2[k][j]

    return result_matrix

def el_mult(matrix1, matrix2):
    n = len(matrix1)
    result_matrix = [[0 for _ in range(n)] for _ in range(n)]

    for i in range(n):
        for j in range(n):
            result_matrix[i][j] = matrix1[i][j] * matrix2[i][j]

    return result_matrix

def np_mult(matrix1_np, matrix2_np):
    return np.dot(matrix1_np, matrix2_np)

def np_el_mult(matrix1_np, matrix2_np):
    return matrix1_np * matrix2_np
```

При помощи timeit засеките время работы функций для n = 100000 (для чистоты
эксперимента засекайте время на одинаковых матрицах).  

```python
mult_time = timeit.timeit(lambda: mult(matrix_A, matrix_B), number=1)
el_mult_time = timeit.timeit(lambda: el_mult(matrix_A, matrix_B), number=1)
np_mult_time = timeit.timeit(lambda: np_mult(matrix_A, matrix_B), number=1)
np_el_mult_time = timeit.timeit(lambda: np_el_mult(matrix_A, matrix_B), number=1)

print("time_mult:", mult_time, "сек")
print("time_el_mult:", el_mult_time, "сек")
print("time_np_mult:", np_mult_time, "сек")
print("time_np_el_mult:", np_el_mult_time, "сек")
```

Во сколько раз numpy версия функций работает быстрее реализаций без numpy?

```python
mult_speed_dif = mult_time/np_mult_time
el_mult_speed_dif = el_mult_time/np_el_mult_time

print("mult быстрее в", mult_speed_dif, "раз")
print("el_mult быстрее в", el_mult_speed_dif, "раз")
```

# 5) В машинном обучении есть модель линейной регрессии, для которой "хорошее" решение считается по следующей формуле. Вычислите результат формулы

```python
X = np.array([[-3, 4, 1], [4, 3, 1]])
y = np.array([[10], [12]])
In = np.eye(3)
l = 0.1

result = (np.linalg.inv(X.T @ X + l * In)) @ X.T @ y

print(result)
```

6. Для каждого наблюдения рассчитайте размах температур (разность максимальной и
минимальной суточных температур) и количество предшествующих ему дней без
осадков (используйте циклы Python и условный оператор).

```python
df['temperature_range'] = np.nan
df['days_without_rain_preceding'] = np.nan


for i, row in enumerate(df.itertuples()):
    if pd.notna(row.min_t) and pd.notna(row.max_t):
        df.loc[i, 'temperature_range'] = row.max_t - row.min_t
    else:
        df.loc[i, 'temperature_range'] = np.nan

    days_without_rain = 0
    for row_index_prev in range(i - 1, -1, -1):
        rainfall_prev = df.loc[row_index_prev, 'rainfall']
        if pd.notna(rainfall_prev) and rainfall_prev == 0:
            days_without_rain += 1 # Увеличиваем счетчик
        else:
            break

    df.loc[i, 'days_without_rain_preceding'] = days_without_rain

print(df.head())
```

8. Используя Seaborn, проведите визуальный анализ и определите кого из топ-20
авторов чаще всего минусуют?

```python
author_views = df.groupby('author')['views'].sum()
top_authors = author_views.nlargest(20)

df_top_authors = df[df['author'].isin(top_authors.index)].copy()

df_top_authors['votes_minus_numeric'] = pd.to_numeric(
    df_top_authors['votes_minus'].astype(str).str.replace(',', '.'),
    errors='coerce'
)

plot_data = pd.merge(
    top_authors.reset_index(),
    df_top_authors.groupby('author')['votes_minus_numeric'].mean().reset_index(),
    on='author',
    how='left'
)

plot_data_sorted = plot_data.sort_values('votes_minus_numeric', ascending=False)

sns.barplot(
    x='votes_minus_numeric',
    y='author',
    data=plot_data_sorted,
    orient='h'
)
plt.title("Топ 20 авторов, отсортированные по количеству минусов")
plt.xlabel("Среднее количество минусов")
plt.ylabel("Автор")
plt.tight_layout()
plt.show()
```
